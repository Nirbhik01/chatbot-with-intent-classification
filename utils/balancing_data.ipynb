```python
import pandas as pd
```

```python
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv('../data/data_with_encoded_result.csv')
```

```python
X = df['vectors_processed_text']
Y = df['result_Encoded']
```

```python
import numpy as np

# Convert string vectors (NumPy array format) to numeric arrays
def parse_vector(vec_str):
    if isinstance(vec_str, str):
        # Remove brackets and split by whitespace, filter empty strings
        cleaned = vec_str.strip('[]').split()
        return np.array([float(x) for x in cleaned if x])
    return vec_str

X_vectors = np.array([parse_vector(vec) for vec in X])

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_vectors, Y)
```

```python
y_resampled.value_counts()
```

```python
new_df = pd.DataFrame({
    "vectors": [vec for vec in X_resampled],
    "result_Encoded": y_resampled
})
```

```python
new_df.to_csv('../data/vectors_and_result.csv',index=False)
```
