```python
import pandas as pd
```

```python
df = pd.read_csv('../data/data_with_vectors.csv')
```

```python
df.head(1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['encoded_result'] = le.fit_transform(df['intent'])
```

```python
df.head(1)
```

```python
result_dict = dict(zip(df['encoded_result'], df['intent']))
```

```python
import json
# Ensure keys are strings for JSON compatibility (optional)
result_dict_strkeys = {str(k): v for k, v in result_dict.items()}
with open('./result_encodings.json', 'w', encoding='utf-8') as f:
    json.dump(result_dict_strkeys, f, indent=4, ensure_ascii=False)
```

```python
df.to_csv('../data/data_with_encoded_result.csv',index=False)
```
