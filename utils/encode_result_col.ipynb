```python
import pandas as pd
```

```python
df = pd.read_csv('../data/data_with_vectors.csv')
```

```python
df.head(1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['encoded_result'] = le.fit_transform(df['intent'])
```

```python
df.head(1)
```

```python
class_to_intent_dict = dict(zip(df['encoded_result'],df['intent']))
```

```python
import json
with open('../data/result_encodings.json', 'w') as f:
    json.dump(class_to_intent_dict, f,indent=4)
```

```python
df.to_csv('../data/data_with_encoded_result.csv',index=False)
```

```python
new_df = df.drop(columns=['text','intent'])
```
