```python
df = pd.read_csv("../data/cleaned_data.csv")
```

```python
from gensim.models import Word2Vec
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
import ast
df["processed_text"] = df["processed_text"].apply(ast.literal_eval)
```

```python
sentences = df["processed_text"].tolist()
```

```python
sentences
```

```python
model = Word2Vec(sentences, vector_size = 200, window = 5, min_count = 1, sg = 1, epochs = 50, negative = 10, workers = 4 )
```

```python
import pickle
```

```python
import os
os.makedirs("../pickle_files", exist_ok=True)

with open("../pickle_files/word2vecmodel.pkl", 'wb') as f:
    pickle.dump(model, f)
```

```python
import numpy as np
```

```python
def vectorize_columns(text):
    # vector = [model.wv[word] for word in text if word in model.wv]
    # return np.mean(vector, axis = 0) if len(vector) > 0 else np.zeros(model.vector_size)
    vecs = np.array([model.wv[w] for w in text if w in model.wv])
    if len(vecs) == 0:
        return np.zeros(model.vector_size * 3)
    return np.concatenate([
        vecs.mean(axis=0),
        vecs.max(axis=0),
        vecs.min(axis=0)
    ])
```

```python
df['vectors_processed_text'] = df['processed_text'].progress_apply(vectorize_columns)
```

```python
df.to_csv("../data/data_with_vectors.csv")
```
